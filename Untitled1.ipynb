```python
# Install required libraries (Uncomment if not installed)
# !pip install gradio transformers

import gradio as gr
from transformers import pipeline
import random
```

```python
# Load text generation model (for captions)
caption_generator = pipeline("text-generation", model="gpt2")
```

```python
# Load sentiment analysis model (for emojis)
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
# Emoji dictionary based on sentiment
emoji_dict = {
    "positive": ["😊", "🌟", "🔥", "💪", "🚀", "✨"],
    "negative": ["😢", "😞", "💔", "😠", "😓"],
    "neutral": ["🙂", "😐", "🧐", "🤔", "😶"]
}
```

```python
# Function to get emojis from sentiment
def get_emojis(text):
    try:
        label = sentiment_pipeline(text)[0]['label'].lower()
        return ''.join(random.sample(emoji_dict.get(label, ["🙂"]), 3))
    except:
        return "🙂🙂🙂"
```

```python
# Function to get hashtags from prompt and platform
def get_hashtags(prompt, platform):
    words = prompt.lower().split()
    tags = ["#" + word.replace(" ", "") for word in words if len(word) > 3]

    platform_tags = {
        "Instagram": ["#instadaily", "#igers", "#picoftheday"],
        "LinkedIn": ["#career", "#leadership", "#networking"],
        "Twitter": ["#tweet", "#trending", "#news"]
    }

    return " ".join(tags[:5] + random.sample(platform_tags.get(platform, []), 2))
```

```python
# Main function to generate output
def generate_post(prompt, platform):
    try:
        result = caption_generator(prompt, max_length=50, num_return_sequences=1, truncation=True)
        caption = result[0].get('generated_text') or result[0].get('text', '')

        emojis = get_emojis(caption)
        hashtags = get_hashtags(prompt, platform)

        return caption.strip(), emojis, hashtags
    except Exception as e:
        return f"Error: {str(e)}", "Error", "Error"
```

```python
# Gradio GUI
interface = gr.Interface(
    fn=generate_post,
    inputs=[
        gr.Textbox(label="Enter keyword or theme", placeholder="e.g., motivational themes"),
        gr.Radio(["Instagram", "LinkedIn", "Twitter"], label="Choose Platform")
    ],
    outputs=[
        gr.Textbox(label="Generated Caption"),
        gr.Textbox(label="Emojis"),
        gr.Textbox(label="Hashtags")
    ],
    title="📱 Social Media Post & Caption Generator",
    description="✨ Create catchy captions, fun emojis, and trending hashtags for your favorite platform using AI.",
    theme="soft",
    allow_flagging="never"
)
```

```python
# Launch the app
interface.launch(share=True)
```
